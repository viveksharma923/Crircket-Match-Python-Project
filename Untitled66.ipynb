#  1 IPL 2022 analysis 

# Most IPL Runs by a Batsman

```python
plt.figure(figsize=(10,8))
data = df.groupby(['batter'])['batsman_run'].sum().sort_values(ascending=False)[:10]
sns.barplot(y=data.index,x=data,orient='h')
plt.xlabel('batsman_run')
plt.ylabel('batter')
plt.show()
```

# Most IPL Century by a Player

```python
runs = df.groupby(['batter','ID'])['batsman_run'].sum()
runs[runs >= 100].droplevel(level=1).groupby('batter').count().sort_values(ascending=False)[:10]
```

# Most IPL Fifty by Player

```python
plt.figure(figsize=(10,8))
runs = df.groupby(['batter','ID'])['batsman_run'].sum()
data = runs[runs >= 50].droplevel(level=1).groupby('batter').count().sort_values(ascending=False)[:10]
sns.barplot(y=data.index,x=data,orient='h')
plt.xlabel('Half-Centuries')
plt.ylabel('Batsman')
plt.show()
```

# Most Sixes in an IPL Inning

```python
plt.figure(figsize=(10,8))
data = df[df['runs_off_bat'] == 6]['striker'].value_counts()[:10]
sns.barplot(y=data.index,x=data,orient='h')
plt.xlabel('Sixes')
plt.ylabel('Batsman')
plt.show()
```

# Most Boundary (4s) hit by a Batsman

```python
plt.figure(figsize=(10,8))
data = df[df['runs_off_bat'] == 4]['striker'].value_counts()[:10]
sns.barplot(y=data.index,x=data,orient='h')
plt.xlabel('Fours')
plt.ylabel('Batsman')
plt.show()
```

# Let's do some analysis on relation between winning the toss and the match

```python
winning_toss_winning_match = df[(df['toss_winner'] == df['match_winner'])]
```

```python
winning_toss_losing_match = df[(df['toss_winner'] != df['match_winner'])]
```

```python
new_df = pd.DataFrame({"Team won the toss":[ winning_toss_winning_match.shape[0],winning_toss_losing_match.shape[0]  ]}, index = ['Won match', 'Lost match']  )
```

```python
new_df.plot( kind = 'bar' , legend = False , title = "Number of time a team won the toss and the match result of that match")
```
